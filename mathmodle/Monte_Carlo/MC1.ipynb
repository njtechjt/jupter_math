蒙特卡洛法——求圆周率

```python
import numpy as np
from matplotlib import pyplot as plt

#初始化参数
p = 10000
r = 1
x0, y0 = 1,1
n = 0  #统计
```

```python
for i in range(p):
    px = np.random.rand()*2  #np.random.rand()产生（0，1）之间的随机数
    py = np.random.rand()*2
    if (px-x0)**2 + (py-y0)**2 < r**2:
        plt.plot(px,py,'.',color='b')
        n += 1
    else:
        plt.plot(px,py,'.',color='r')
plt.axis('equal')
#设置绘图窗口
# plt.figure()  #打开一个画板  有几个就打开几个画板
plt.title('Monte Carlo Simvlation for pi')
plt.xlabel('x')
plt.ylabel('y',rotation=0)
plt.show()
```

```python
#记算pi的值
s = (n/p)*4
print('pi~=:',s)
```
