# 泰勒展开
计算麦克劳林级数
series_expansion = series(y, x, 0, k)
# 关键参数
参数	说明
y	待展开的表达式
x	展开变量
0	展开点（此处为 x=0）
k	展开阶数（最高次项为 x^(k-1)）

```python
from pylab import rc
from sympy import *
from sympy.plotting import *

rc('font',size=16); rc('text',usetex=True)
x = symbols('x')
y = sin(x)
for k in range(3,8,2):
    print(series(y,x,0,k))
plot(y,series(y,x,0,3).removeO(),series(y,x,0,5).removeO(),series(y,x,0,7).removeO(), (x,0,3), xlabel='$x$', ylabel='$y$')  #绘制同阶数的泰勒展开（移除余项） 
#(x, 0, 2) 表示x轴范围
```
