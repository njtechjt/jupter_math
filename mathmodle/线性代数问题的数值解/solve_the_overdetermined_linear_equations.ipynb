# 求超定线性方程组的最小二乘解
建模中经常使用线性最小二乘解，实际上就是求超定线性方程组(未知数个数少，方程个数多)的最小二乘解，可以使用pinv()求超定线性方程组的最小二乘解，也可以使用numpy.linalg模块的lstsq()函数求解
####
    相当于给定x的观测值和y的观测值，拟合经验函数y = mx+c

```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt

x = np.array([0,1,2,3])
y = np.array([-1,0.2,0.9,2.1])
A = np.c_[x, np.ones_like(x)]
m, c = la.lstsq(A,y,rcond=None)[0]  #我们通常只需要第一个元素 [0]，即最小二乘解。
print(m,c)
plt.rc('font',size=16)
plt.plot(x,y,'o', label='原始数据', markersize=5)
plt.plot(x, m*x+c,'r',label='拟合直线')
plt.rc('font', family='SimHei')  #用来正常显示中文标签
# plt.rc('axes', unicode_minus=False)  #用来正常显示负号
plt.legend()
plt.show()
```

```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt

t = np.arange(8)
y = np.array([27.0,26.8,26.5,26.3,26.1,25.7,25.3,24.8])
A = np.c_[t, np.ones_like(t)]
ab = la.lstsq(A,y,rcond=None)[0]
print(ab)

plt.rc('font', size=16)
plt.plot(t,y,'o', label='Original data', markersize=5)
plt.plot(t, A.dot(ab), 'r', label='Fitted line')
plt.legend(); plt.show()
```
