# 泰勒级数

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 101)
plt.plot(x, np.sin(x), '*-')
#自定义sin(x)的泰勒级数
def fac(n):
    if n<1:
        return 1
    else:
        return n*fac(n-1)
def item(n,x):
    return (-1)**n*x**(2*n+1)/fac(2*n+1)
def mysin(n, x):
    if n<0:
        return 0
    else:
        return mysin(n-1,x)+item(n,x)
str = ['v-', 'H--', '-.']
for n in np.arange(1,4):
    plt.plot(x, mysin(2*n-1,x), str[n-1])
plt.legend(['sin(x)','n=1','n=3','n=5'])
plt.show()
```
